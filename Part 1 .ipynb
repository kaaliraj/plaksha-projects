```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
google_data = pd.read_csv("google.txt",sep = "\t")
ny_data = pd.read_csv("ny.txt",sep = "\t" )
yahoo_data = pd.read_csv("yahoo.txt",sep = "\t")
```

```python
#Creating plot with dataset_1 
fig, ax = plt.subplots(figsize=(10,5)) 

ax.set_xlabel('Date (MJD)')  
ax.set_ylabel('Value (Dollars)', color = 'purple')  
plot_1 = ax.plot(yahoo_data['Modified Julian Date'], yahoo_data['Stock Value'], color = 'purple', label ='Yahoo! Stock Value')  
plot_2 = ax.plot(google_data['Modified Julian Date'], google_data['Stock Value'], color = 'green', label ='Google Stock Value')  


ax_twin = ax.twinx()  
ax_twin.set_ylim(-150,100)  
ax_twin.set_ylabel('Temperature (°F)', color = 'blue')  
plot_3 = ax_twin.plot(ny_data['Modified Julian Date'], ny_data['Max Temperature'],'g--' ,color = 'blue', label ='NY Mon. High Temp')  
ax_twin.tick_params(axis ='y', labelcolor = 'blue')  

plot= plot_1 + plot_2 + plot_3
labels = [i.get_label() for i in plot]


plt.title('New York Temperature, Google, and Yahoo!', fontweight="bold")  
plt.legend(plot, labels, loc='center left', frameon=False) 

plt.show()
```
